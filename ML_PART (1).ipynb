```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import folium
```

```python
df = pd.read_excel('FULLDATA.xlsx')
df
```

```python
def count_skills(row):
    if pd.notnull(row):
        return len(row.split(','))
    else:
        return 0

df['Count_of_Skills'] = df['Skills'].apply(count_skills)

df[['Skills', 'Count_of_Skills']]
```

```python
import pandas as pd

df['Salary_KZT'] = ((df['Min_Salary_KZT'] + df['Max_Salary_KZT']) / 2).astype(int)

df[['Salary_KZT']]
```

```python
scaler = MinMaxScaler()

df['Normalized_Salary_KZT'] = scaler.fit_transform(df[['Salary_KZT']].values)

df[['Salary_KZT', 'Normalized_Salary_KZT']]
```

```python
min_exp_min = df['Min_Experience'].min()
min_exp_max = df['Min_Experience'].max()

df['Normalized_Min_Experience'] = 1 - ((df['Min_Experience'] - min_exp_min) / (min_exp_max - min_exp_min))

df[['Min_Experience', 'Normalized_Min_Experience']]
```

```python
min_skl_min = df['Count_of_Skills'].min()
min_skl_max = df['Count_of_Skills'].max()

df['Normalized_Count_of_Skills'] = 1 - ((df['Count_of_Skills'] - min_skl_min) / (min_skl_max - min_skl_min))

df[['Count_of_Skills', 'Normalized_Count_of_Skills']]
```

```python
df['Job_rate'] = 40 * df['Normalized_Salary_KZT'] + 25 * df['Normalized_Min_Experience'] + 35 * df['Normalized_Count_of_Skills']
df
```
