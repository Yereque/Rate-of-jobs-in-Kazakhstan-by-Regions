```python
multiple_cities = df['Address'].str.contains('|'.join(['Алматы', 'Отеген']), case=False, na=False)
df.loc[multiple_cities, 'Address'] = 'Алматы'

kyzylorda = df['Address'].str.contains('Кызылорд', case=False, na=False)
df.loc[kyzylorda, 'Address'] = 'Кызылорда'

Сатпаев = df['Address'].str.contains('Сатпаев', case=False, na=False)
df.loc[Сатпаев, 'Address'] = 'Жезказган'

astana = df['Address'].str.contains('Щучинск', case=False, na=False)
df.loc[astana, 'Address'] = 'Астана'

city = df['Address'].unique()
len(city)
```

## Experience Standardization: Enhancing DF with Min and Max Experience Columns

```python
def process_experience(text):
    import re
    min_experience_match = re.search(r'от (\d+)', text)
    max_experience_match = re.search(r'до (\d+)', text)
    more_experience_match = re.search(r'более (\d+)', text)
    no_experience_match = re.search(r'Без опыта', text)
    
    min_experience = int(min_experience_match.group(1)) if min_experience_match else None
    max_experience = int(max_experience_match.group(1)) if max_experience_match else None
    more_experience = int(more_experience_match.group(1)) if more_experience_match else None
    
    if no_experience_match:
        min_experience = 0
        max_experience = 0
    elif more_experience:
        min_experience = more_experience
        max_experience = more_experience
        
    return min_experience, max_experience

df['Min_Experience'], df['Max_Experience'] = zip(*df['Experience'].apply(process_experience))

df.to_excel('df_with_exp.xlsx', index=False)
```

## Text Cleaning: Removing Newlines and Standardizing 'Description' in DataFrame

```python
def clean_description(text):
    if isinstance(text, str):
        cleaned_text = text.replace('/n', '').strip().capitalize()
        return cleaned_text
    else:
        return text  

df['Description'] = df['Description'].apply(clean_description)

df.to_excel('without_N_and_exp.xlsx', index=False)
df
```

## Salary Conversion and Standardization: Transforming Currency Values to KZT and Adjusting Ranges in df

```python
df['Currency'] = df['Salary'].str[-1]

currency_counts = df['Currency'].value_counts()

print("Количество встреч каждой валюты:")
print(currency_counts)
df
```

```python
exchange_rates_data = {'Currency': ['₸', '₽', '$', '€', '₾'],
                       'Exchange_Rate': [1, 4.98, 458, 505, 162]}
exchange_rates_df = pd.DataFrame(exchange_rates_data)

def process_salary(salary_string):
    if isinstance(salary_string, str):
        numbers = [int(s) for s in salary_string.split() if s.isdigit()]
        
        # Initialize variables with None
        min_salary = max_salary = None
        
        if 'от' in salary_string:
            min_salary = numbers[0]
            max_salary = min_salary * 2
        elif 'до' in salary_string:
            max_salary = numbers[0]
            min_salary = max_salary / 2
        elif '-' in salary_string:
            if len(numbers) > 2:
                min_salary = int(numbers[0])
                max_salary = int(numbers[2])
        elif '–' in salary_string:
            if len(numbers) > 2:
                min_salary = int(numbers[0])
                max_salary = int(numbers[2])
        
        currency = salary_string.split()[-1]
        if currency not in exchange_rates_df['Currency'].values:
            return None, None
        
        exchange_rate = exchange_rates_df.loc[exchange_rates_df['Currency'] == currency, 'Exchange_Rate'].values[0]
        min_salary_kzt = min_salary * exchange_rate if min_salary is not None else None
        max_salary_kzt = max_salary * exchange_rate if max_salary is not None else None
        
        return min_salary_kzt, max_salary_kzt
    else:
        return salary_string, salary_string

df[['Min_Salary_KZT', 'Max_Salary_KZT']] = pd.DataFrame(df['Salary'].apply(process_salary).to_list(), index=df.index)
pd.set_option('display.float_format', lambda x: f'{x:.0f}' if x == x else '')
print("Зарплаты в тенге:")
print(df[['Salary', 'Min_Salary_KZT', 'Max_Salary_KZT']])
```

## Final Data Processing and Imputation: Saving, Handling Missing Values, and Extracting Most Common Skills in DF

```python
df.isna().sum()
```

```python
df['Min_Salary_KZT'] = df['Min_Salary_KZT'].fillna(df['Min_Salary_KZT'].mean())
df['Max_Salary_KZT'] = df['Max_Salary_KZT'].fillna(df['Max_Salary_KZT'].mean())
```

```python
mode_value = df['Company'].mode()[0]
df['Company'].fillna(mode_value, inplace=True)
```

```python
all_skills = []
for skills in df['Skills'].dropna():
    all_skills.extend(skills.split(','))

most_common_skills = pd.Series(all_skills).value_counts().head(5).index.tolist()

df['Skills'] = df['Skills'].apply(lambda x: most_common_skills if pd.isnull(x) else x.split(','))
```

```python
df['Skills'] = df['Skills'].apply(lambda x: ', '.join(x) if isinstance(x, list) else x)
```

## Dynamics of Salaries Depending on Work Experience

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

experience_salary_trend = df.groupby(['Min_Experience', 'Max_Experience']).agg({'Min_Salary_KZT': 'mean', 'Max_Salary_KZT': 'mean'}).reset_index()

sns.lineplot(x='Min_Experience', y='Min_Salary_KZT', data=experience_salary_trend,
             linestyle='-.', marker='^', markersize=20, color='purple', label='Minimum Salary')

sns.lineplot(x='Max_Experience', y='Max_Salary_KZT', data=experience_salary_trend,
             linestyle='-', marker='o', markersize=8, color='blue', label='Maximum Salary')


plt.title('Salary Trends Based on Experience', fontsize=16)
plt.xlabel('Experience', fontsize=14)
plt.ylabel('Salary (KZT)', fontsize=14)
plt.legend()

plt.show()
```

## Salary Range in Top 400 Companies: Scatter Chart

```python
import matplotlib.pyplot as plt

top_50_companies = df['Company'].value_counts().head(400).index.tolist()

company_salaries = {}
for company in top_50_companies:
    max_salary = df[df['Company'] == company]['Max_Salary_KZT'].max()
    min_salary = df[df['Company'] == company]['Min_Salary_KZT'].min()
    company_salaries[company] = {'Max_Salary_KZT': max_salary, 'Min_Salary_KZT': min_salary}

max_salaries = [company_salaries[company]['Max_Salary_KZT'] for company in top_50_companies]
min_salaries = [company_salaries[company]['Min_Salary_KZT'] for company in top_50_companies]

plt.figure(figsize=(10, 6))
plt.scatter(min_salaries, max_salaries, alpha=0.7)
plt.title('Scatter Plot of Minimum vs Maximum Salaries for Top 400 Companies')
plt.xlabel('Minimum Salary (KZT)')
plt.ylabel('Maximum Salary (KZT)')
plt.grid(True)
plt.show()
```
